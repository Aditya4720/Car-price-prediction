```python
import pandas as pd
a=pd.read_csv("Car data.csv")
a
```

```python
a.drop(columns=["torque"],inplace=True)
```

```python
a.isnull().sum()
```

```python
a.dropna(inplace=True)
```

```python
a.drop_duplicates(inplace=True)
```

```python
a.info()
```

```python
a['name'] = a['name'].apply(lambda x: x.split()[0])
```

```python
a['max_power'] = a['max_power'].apply(lambda x: x.split()[0])
```

```python
# Replace 'bhp' and strip whitespaces
a['max_power'] = a['max_power'].str.replace('bhp', '', regex=False).str.strip()
    
# Convert to numeric (invalid entries like empty strings will become NaN)
a['max_power'] = pd.to_numeric(a['max_power'], errors='coerce')

# Optionally, fill or drop missing values
a['max_power'].fillna(a['max_power'].mean(), inplace=True)
```

```python
# Step 1: Remove 'CC' and strip spaces
a['engine'] = a['engine'].str.replace('CC', '', regex=False).str.strip()

# Step 2: Convert to float
a['engine'] = pd.to_numeric(a['engine'], errors='coerce')

# Step 3: Fill missing values with mean (optional)
a['engine'].fillna(a['engine'].mean(), inplace=True)
```

```python

a['mileage'] = a['mileage'].str.replace('kmpl', '', regex=False)
a['mileage'] = a['mileage'].str.replace('km/kg', '', regex=False)
a['mileage'] = a['mileage'].str.strip()


a['mileage'] = pd.to_numeric(a['mileage'], errors='coerce')


a['mileage'].fillna(a['mileage'].mean(), inplace=True)
```

```python
a
```

```python
a["name"].unique()
```

```python
import pickle

# Suppose you encoded these columns
from sklearn.preprocessing import LabelEncoder

le_name = LabelEncoder()
le_fuel = LabelEncoder()
le_seller_type = LabelEncoder()
le_transmission = LabelEncoder()
le_owner = LabelEncoder()

# Fit them to data
a["name"]=le_name.fit_transform(a['name'])
a['fuel']=le_fuel.fit_transform(a['fuel'])
a['seller_type']=le_seller_type.fit_transform(a['seller_type'])
a['transmission']=le_transmission.fit_transform(a['transmission'])
a['owner']=le_owner.fit_transform(a['owner'])
# Save them
pickle.dump(le_name, open('name_encoding.pkl', 'wb'))
pickle.dump(le_fuel, open('fuel_encoding.pkl', 'wb'))
pickle.dump(le_seller_type, open('seller_encoding.pkl', 'wb'))
pickle.dump(le_transmission, open('transmission_encoding.pkl', 'wb'))
pickle.dump(le_owner, open('owner_encoding.pkl', 'wb'))
```

```python
x = a.drop('selling_price', axis=1)
y = a['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Create and train the model
model = XGBRegressor(
    n_estimators=200,
    max_depth=10,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(x_train, y_train)

# Predict on train and test
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Train Metrics
print("ðŸ”¹ Train RÂ² Score:", r2_score(y_train, y_train_pred))
print("ðŸ”¹ Train MAE:", mean_absolute_error(y_train, y_train_pred))
print("ðŸ”¹ Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))

# Test Metrics
print("ðŸ”¹ Test RÂ² Score:", r2_score(y_test, y_test_pred))
print("ðŸ”¹ Test MAE:", mean_absolute_error(y_test, y_test_pred))
print("ðŸ”¹ Test RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
import pickle

with open('xgboost_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
